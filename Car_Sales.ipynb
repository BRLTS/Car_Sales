```python
# Mengimpor modul pandas ke environment
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/new2017.csv')
```

```python
df
```

```python
df.drop(["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC", "SHARE"],axis=1,inplace=True)
```

```python
df
```

```python
# prompt: Menggunakan df DataFrame: check PROD data type

df['PROD'] = pd.to_numeric(df['PROD'], errors='coerce')
```

```python
df = df.sort_values('PROD', ascending=False)
```

```python
df
```

```python
# prompt: Menggunakan df DataFrame: remove cells with NaN values

df.dropna(subset=['PROD'], inplace=True)
```

```python
df
```

```python
df = df[0:11]
```

```python
df
```

```python
df = df.sort_values('BRAND')
```

```python
x = df['BRAND']
y = df['PROD']

plt.bar(x, y, label="Production", color='b')
plt.plot()

plt.xlabel("Car Brands")
plt.ylabel("Total Production")
plt.title("Indonesia Car Sales Production 2017")
plt.rcParams["figure.figsize"] = (20, 10)

plt.legend()
plt.show()
```
